```python
import pandas as pd
import os
import json
import itertools
```

```python
features = {}
features_all = {}

for exp in range(1, len(os.listdir("./Experiments"))+1):

    with open(f"./Experiments/{exp}/metadata.json", 'r') as f:
        metadata = json.load(f)

    feature_issue = metadata['feature_chg_idx']
    feature_issue_count_all = []
    feature_issue_count = []

    all_dec_path = f"./Experiments/{exp}/all_decisions"
    target_nds_path = f"./Experiments/{exp}/target_node_decisions"
    num_files = len(os.listdir(all_dec_path))-1

    for i in range(1, num_files+1):

        data_all = pd.read_csv(f"{all_dec_path}/change_{i}_features.csv", index_col=[0,1])
        data = pd.read_csv(f"{target_nds_path}/change_{i}_features.csv", index_col=[0,1])

        if feature_issue in data_all['Feature name'].values[:3]:
            feature_issue_count_all.append(1)

        else:
            feature_issue_count_all.append(0)
        
        if feature_issue in data['Feature name'].values[:3]:
            feature_issue_count.append(1)

        else:
            feature_issue_count.append(0)
    try:
        features_all[feature_issue] = [sum(x) for x in 
            itertools.zip_longest(*[features_all[feature_issue], feature_issue_count_all], fillvalue=0)]

    except KeyError:
        features_all[feature_issue] = feature_issue_count_all
    
    try:
        features[feature_issue] = [sum(x) for x in 
            itertools.zip_longest(*[features[feature_issue], feature_issue_count], fillvalue=0)]

    except KeyError:
        features[feature_issue] = feature_issue_count    
    
```

```python
df = pd.DataFrame.from_dict(features, orient='index').fillna(0).astype('int64')
df.columns = df.columns+1
df
```

```python
df_all = pd.DataFrame.from_dict(features_all, orient='index').fillna(0).astype('int64')
df_all.columns = df_all.columns+1
df_all
```
